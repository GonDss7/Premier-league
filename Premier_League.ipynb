# Task
Analiza qué otros factores, además de los goles, podrían influir en el salario de los jugadores.

## Seleccionar factores potenciales

### Subtask:
Identificar columnas en los DataFrames `player_stats` y `player_possession` que podrían estar relacionadas con el salario de un jugador (excluyendo `player_id`, `Player`, `Team`, `Nation`, `Born`).


**Reasoning**:
Identify relevant columns from `player_stats` and `player_possession` dataframes, excluding the specified columns, and combine the lists.



```python
# Identify potential factor columns from player_stats
player_stats_potential_factors = [col for col in player_stats.columns if col not in ['player_id', 'Player', 'Team', 'Nation', 'Born']]

# Identify potential factor columns from player_possession
player_possession_potential_factors = [col for col in player_possession.columns if col not in ['player_id', 'Player', 'Team']]

# Combine the lists of potential factor columns
all_potential_factors = list(set(player_stats_potential_factors + player_possession_potential_factors))

print("Potential factors influencing player salary:")
print(all_potential_factors)
```

## Fusionar datos

### Subtask:
Asegurarse de tener un DataFrame combinado que contenga el salario anual y las columnas de los factores potenciales para cada jugador. Ya tenemos `merged_player_data` que incluye datos de `player_stats` y `player_salaries`. Necesitaremos fusionar `player_possession` también.


**Reasoning**:
Merge the two dataframes to combine player stats, possession data, and salary information.



```python
combined_player_data = pd.merge(merged_player_data, player_possession, on='player_id', how='inner', suffixes=('_stats', '_possession'))
display(combined_player_data.head())
```

## Calcular correlaciones

### Subtask:
Calcular la matriz de correlación entre el salario anual y las columnas de los factores potenciales seleccionados.


**Reasoning**:
Select the relevant columns for correlation analysis from the `combined_player_data` DataFrame, including 'Annual' and the potential factor columns identified previously, while excluding identification columns and duplicated columns with suffixes. Then, calculate the correlation matrix and display the correlation of 'Annual' with the other selected columns, sorted in descending order.



**Reasoning**:
The error message "ValueError: could not convert string to float: 'DF'" indicates that there is a non-numeric column being included in the correlation calculation. The column 'Position' contains string values like 'DF', 'MF', 'FW', etc., which are not suitable for correlation. I need to exclude this column before calculating the correlation matrix.



```python
# Select 'Annual' and potential factor columns from the combined DataFrame,
# excluding original 'Player' and 'Team' columns from player_possession,
# duplicated columns with suffixes other than the ones we want to correlate,
# and the 'Position' column.
potential_factors_with_salary = ['Annual'] + [
    col for col in combined_player_data.columns
    if (col in all_potential_factors or col == 'Annual')
    and '_possession' not in col
    and '_stats' not in col
    and col != 'Position' # Exclude 'Position' column
]

# Ensure 'Annual' is in the list and remove any duplicates if they exist
potential_factors_with_salary = list(set(potential_factors_with_salary))

# Calculate the correlation matrix
correlation_matrix = combined_player_data[potential_factors_with_salary].corr()

# Get the correlation with 'Annual' and sort
annual_correlation = correlation_matrix['Annual'].sort_values(ascending=False)

print("Correlation with Annual Salary:")
display(annual_correlation)
```

Create scatter plots to visualize the relationship between 'Annual' salary and the selected factors with the highest correlation, including 'Progressive_passes', 'Age', 'Received', and 'Goals'.

```python
from sklearn.preprocessing import MinMaxScaler

# Select the specified performance metrics and salary
performance_cols = ['Assists', 'Goals', 'Carries', 'Touches', 'Total_distance_carried']
salary_col = 'Annual'

# Select the relevant data
performance_data = combined_player_data[performance_cols]
salary_data = combined_player_data[salary_col]

# Normalize the performance metrics
scaler = MinMaxScaler()
normalized_performance = scaler.fit_transform(performance_data)
normalized_performance_df = pd.DataFrame(normalized_performance, columns=[f'{col}_normalized' for col in performance_cols])

# Calculate a combined performance score (e.g., sum of normalized metrics)
normalized_performance_df['Performance_Score'] = normalized_performance_df.sum(axis=1)

# Add player information and salary back to the performance score DataFrame
player_revelation_data = combined_player_data[['player_id', 'Player_x', 'Team_x']].copy()
player_revelation_data['Annual_Salary'] = salary_data

# Concatenate the normalized performance scores and the original data
player_revelation_data = pd.concat([player_revelation_data, normalized_performance_df], axis=1)

# Define "low salary" - for example, below the median annual salary
median_annual_salary = player_revelation_data['Annual_Salary'].median()
low_salary_players = player_revelation_data[player_revelation_data['Annual_Salary'] <= median_annual_salary]

# Sort players with low salary by their performance score in descending order
revelations = low_salary_players.sort_values(by='Performance_Score', ascending=False)

print("Top 10 Players with High Performance (based on selected metrics) and Low Annual Salary:")
display(revelations[['Player_x', 'Team_x', 'Annual_Salary', 'Performance_Score'] + [f'{col}_normalized' for col in performance_cols]].head(10))
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=player_revelation_data, x='Annual_Salary', y='Performance_Score', alpha=0.7)

# Highlight the top 10 revelations
sns.scatterplot(data=revelations.head(10), x='Annual_Salary', y='Performance_Score', color='red', s=100, label='Top 10 Revelations')

# Add annotations for the top 10 players
for i, row in revelations.head(10).iterrows():
    plt.text(row['Annual_Salary'], row['Performance_Score'], row['Player_x'], fontsize=9, ha='left', va='bottom')

plt.title('Annual Salary vs. Performance Score (Highlighting Top 10 Revelations)')
plt.xlabel('Annual Salary')
plt.ylabel('Performance Score')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
